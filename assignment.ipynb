# Train Test split


```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,0:4]  
y = df.iloc[:,-1]    
```

```python
X.head(5)
```

```python
y.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=3)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Activation, Flatten
```

# Model


```python
nnmodel = Sequential()
#First hidden layer
nnmodel.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu', input_dim = 4))
#Going to use sigmoid instead of softmax as clearly mentioned in the instruction the formulea is in the form of y = c+mx(c =K0,mx= K1x)
nnmodel.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
nnmodel.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Feeding data to the ann architecture
nnhistory = nnmodel.fit(X_train, y_train, batch_size = 10, epochs = 100)

Score, Accuracy = nnmodel.evaluate(X_train, y_train,
                            batch_size=10)
print('Train score:', Score)
print('Train accuracy:', Accuracy)

# Predicting the Test set results
y_pred = nnmodel.predict(X_test)
y_pred = (y_pred > 0.5)

print('*'*100)
Score, Accuracy = nnmodel.evaluate(X_test, y_test,
                            batch_size=10)
print('Test score:', Score)
print('Test accuracy:', Accuracy)
```

```python
nnmodel = Sequential()
nnmodel.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu', input_dim = 4))
nnmodel.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
nnmodel.summary()
```

```python
nnmodel.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['mae', 'acc'])

nnhistory =nnmodel.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size = 10, epochs = 100,verbose = 0)
```

```python
plt.figure(figsize=(8, 4))  

# summarize history for accuracy
plt.plot(nnhistory.history['acc'])
plt.plot(nnhistory.history['val_acc'])
plt.title('ANN_model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test' ], loc='upper left')
plt.show()
```
